```python
import numpy as np
import matplotlib.pyplot as plt
import numba

#performes central differences differentiation in theta direction
@numba.njit
def diff(array, dtheta):
    return (array[2:] - array[:-2])/(dtheta*2)

@numba.njit
def cylindrical_to_spherical(values, theta_grid, a):
    return values / np.cos(theta_grid)**a

@numba.njit
def pad(values, amount, constant=True):
    res = np.empty(len(values) + 2*amount)
    res[amount:-amount] = values
    if constant:
        res[:amount] = values[0]
        res[-amount:] = values[-1]
    else:
        for i in range(amount):
            res[amount-i-1] = (i+1) * values[0] - values[1]
        res[:amount] = values[0]
        res[-amount:] = values[-1]
    return res
        

# represents a scale invariant vector field, and provides assosiated vector operations
# ie a field such that f(r,theta,phi)=val(theta)*(r/r_0)**a
# at the point r=r_0 in units of r_0=1
#@numba.experimental.jitclass([("v_r", numba.float64[:]), ("v_theta", numba.float64[:]), ("v_phi", numba.float64[:]), ("a", numba.float64), ("dtheta", numba.float64), ("theta_grid", numba.float64[:])])
class VectorField:
    def __init__(self, v_r, v_theta, v_phi, a, theta_grid, prescription="spherical"):
        #assert (prescription in ["theta", "z"]), f"val must be prescibed as either Q(z/R) or Q(theta). was given Q({prescription})"
        #theta_diff = np.diff(theta_grid)
        #assert (np.max(theta_diff) - np.min(theta_diff)) < 1e-6, "all theta values must be identical"
        #else:
        #    self.val = np.zeros_like(theta_grid) + val
        if prescription == "spherical":
            self.v_r = v_r
            self.v_theta = v_theta
            self.v_phi = v_phi
        elif prescription == "cylindrical":
            self.v_r = cylindrical_to_spherical(v_r, theta_grid, a) 
            self.v_theta = cylindrical_to_spherical(v_theta, theta_grid, a) 
            self.v_phi = cylindrical_to_spherical(v_phi, theta_grid, a) 
        else:
            raise NotImplementedError(f"prescription not implemented")

        self.a = a
        self.theta_grid = theta_grid
        self.dtheta = theta_grid[1] - theta_grid[0]
            
    def __repr__ (f):
        return f"({f.v_r}\\hat{{r}}\n+{f.v_theta}\\hat{{theta}}\n+{f.v_phi}\\hat{{phi}})\n* (r/r_0)^{f.a}"
        
    # method for extracting a part of the field. Usually usefull for cutting off the edges
    def __getitem__(self, key):
        return VectorField(self.v_r[key], self.v_theta[key], self.v_phi[key], self.a, self.theta_grid[key])
            
    def div(f):
        f_tilde = (f.a+2)*f.v_r[1:-1] \
                + 1/np.sin(f.theta_grid[1:-1])*diff(np.sin(f.theta_grid) * f.v_theta, f.dtheta)
        return ScalarField(f_tilde, f.a-1, f.theta_grid[1:-1])
    
    #Implements directional deriv (g \cdot \nabla f) as f.direction_deriv(g)
    def direction_deriv(f, g): 
        cot = 1/np.tan(f.theta_grid[1:-1])
        g_new = g[1:-1]
        res_r     = (g_new * ScalarField(f.v_r, f.a, f.theta_grid).grad()).v     \
                    - g_new.v_theta * f.v_theta[1:-1] - g_new.v_phi*f.v_phi[1:-1]
        res_theta = (g_new * ScalarField(f.v_theta, f.a, f.theta_grid).grad()).v \
                    - g_new.v_theta * f.v_r[1:-1] - g_new.v_phi*f.v_phi[1:-1]*cot
        res_phi   = (g_new * ScalarField(f.v_phi, f.a, f.theta_grid).grad()).v   \
                    - g_new.v_phi  *  f.v_r[1:-1] + g_new.v_phi*f.v_theta[1:-1]*cot
        return VectorField(res_r, res_theta, res_phi, g.a+f.a-1, g_new.theta_grid)
    
    def curl(f):
        f_new = f[1:-1]
        f_tilde = [1/np.sin(f_new.theta_grid) * diff(f.v_phi*np.sin(f.theta_grid), f.dtheta),
                   -(f.a+1)*f_new.v_phi,
                   +(f.a+1)*f_new.v_theta - diff(f.v_r, f.dtheta)]
        return VectorField(*f_tilde, f.a-1, f_new.theta_grid)
    
    # vector field * vector field = dot product
    # vector field * scalar field = scalar product
    # scalar field * scalar field = normal multi
    def __mul__(f, g):
        if type(g) is VectorField:
            #assert np.all(np.isclose(f.theta_grid, g.theta_grid)), "need to be defined on the same domain"
            return ScalarField(f.v_r*g.v_r + f.v_theta*g.v_theta + f.v_phi*g.v_phi, f.a + g.a, f.theta_grid)
        elif type(g) is ScalarField:
            return VectorField(f.v_r*g.v, f.v_theta*g.v, f.v_phi*g.v, f.a+g.a, f.theta_grid)
        else: #scalar
            return VectorField(f.v_r*g, f.v_theta*g, f.v_phi*g, f.a, f.theta_grid)

    # vector field / scalar field 
    # scalar field / float 
    def __truediv__(f, g):
        if type(g) is ScalarField:
            #assert np.all(np.isclose(f.theta_grid, g.theta_grid)), "need to be defined on the same domain"
            return VectorField(f.v_r/g.v_r, f.v_theta/g.v_theta, f.v_phi/g.v_phi,
                               f.a - g.a, f.theta_grid)
        else:
            raise NotImplementedError(f"type of g not implemented")
    
    def cross(f, g):
        #assert np.all(np.isclose(f.theta_grid, g.theta_grid)), "need to be defined on the same domain"
        #assert f.val.shape == g.val.shape and f.val.ndim == 2, "need to both be vector fields"
        return VectorField(f.v_theta*g.v_phi-f.v_phi*f.v_theta,
                           f.v_phi*g.v_r-f.v_r*g.v_phi,
                           f.v_r*g.v_theta-f.v_theta*f.v_r, 
                           f.a + g.a, f.theta_grid)
    
    def __add__(f, g):
        #assert np.all(np.isclose(f.theta_grid, g.theta_grid)), "need to be defined on the same domain"
        #assert f.val.shape == g.val.shape,  f"need to have the same shape. has {f.val.shape} and {g.val.shape}"
        #assert f.a == g.a, f"need to have the same power index. has {f.a} and {g.a}"
        return VectorField(f.v_r+g.v_r, f.v_theta+g.v_theta, f.v_phi+g.v_phi, f.a, f.theta_grid)
    
    def __sub__(f, g):
        return VectorField(f.v_r-g.v_r, f.v_theta-g.v_theta, f.v_phi-g.v_phi, f.a, f.theta_grid)
    
    def __neg__(f): 
        return VectorField(-f.v_r, -f.v_theta, -f.v_phi, f.a, f.theta_grid)

    def __pow__(f, scalar): 
        return VectorField(f.v_r**scalar, f.v_theta**scalar, f.v_phi**scalar, f.a*scalar, f.theta_grid)
    
    def unit(f):
        #assert f.val.ndim == 2, "need to be vector field"
        norm = f.norm()
        return VectorField(f.v_r/norm, f.v_theta/norm, f.v_phi/norm, 0, f.theta_grid)
    
    def norm(f):
        return (f.v_r**2+f.v_theta**2+f.v_phi**2)**0.5

# represents a scale invariant vector field, and provides assosiated vector operations
# ie a field such that f(r,theta,phi)=val(theta)*(r/r_0)**a
# at the point r=r_0 in units of r_0=1
#@numba.experimental.jitclass([("v", numba.float64[:]), ("a", numba.float64), ("dtheta", numba.float64), ("theta_grid", numba.float64[:])])
class ScalarField:
    def __init__(self, v, a, theta_grid, prescription="spherical"):
        #assert (prescription in ["theta", "z"]), f"val must be prescibed as either Q(z/R) or Q(theta). was given Q({prescription})"
        #theta_diff = np.diff(theta_grid)
        #assert (np.max(theta_diff) - np.min(theta_diff)) < 1e-6, "all theta values must be identical"
        #else:
        #    self.val = np.zeros_like(theta_grid) + val
        v_ = np.zeros_like(theta_grid) + v
        if prescription == "spherical":
            self.v = v_
        elif prescription == "cylindrical":
            self.v = cylindrical_to_spherical(v_, theta_grid, a) 
        else:
            raise NotImplementedError(f"prescription not implemented")
        self.a = a
        self.theta_grid = theta_grid
        self.dtheta = theta_grid[1] - theta_grid[0]
            
    def __repr__ (f):
        return f"{f.v} * (r/r_0)^{f.a}"
        
    # method for extracting a part of the field. Usually usefull for cutting off the edges
    def __getitem__(self, key):
        return ScalarField(self.v[key], self.a, self.theta_grid[key])
        
    def grad(f):
        f_tilde = ([f.a*f.v[1:-1], diff(f.v, f.dtheta), np.zeros(len(f.v)-2)])
        return VectorField(*f_tilde, f.a-1, f.theta_grid[1:-1])
    
    # vector field * vector field = dot product
    # vector field * scalar field = scalar product
    # scalar field * scalar field = normal multi
    def __mul__(f, g):
        if type(g) is VectorField:
            return g*f
        elif type(g) is ScalarField:
            return ScalarField(f.v*g.v, f.a+g.a, f.theta_grid)
        else: #scalar
            return VectorField(f.v*g, f.a, f.theta_grid)

    # vector field / scalar field 
    # scalar field / float 
    def __truediv__(f, g):
        if type(g) is ScalarField:
            #assert np.all(np.isclose(f.theta_grid, g.theta_grid)), "need to be defined on the same domain"
            return ScalarField(f.v/g, f.a - g.a, f.theta_grid)
        if isinstance(g, (int,float)):
            #assert np.all(np.isclose(f.theta_grid, g.theta_grid)), "need to be defined on the same domain"
            return ScalarField(f.v/g, f.a, f.theta_grid)
        else:
            raise NotImplementedError(f"type of g not implemented")

    def __add__(f, g):
        #assert np.all(np.isclose(f.theta_grid, g.theta_grid)), "need to be defined on the same domain"
        #assert f.val.shape == g.val.shape,  f"need to have the same shape. has {f.val.shape} and {g.val.shape}"
        #assert f.a == g.a, f"need to have the same power index. has {f.a} and {g.a}"
        return ScalarField(f.v+g.v, f.a, f.theta_grid)
    
    def __sub__(f, g):
        return ScalarField(f.v-g.v, f.a, f.theta_grid)
    
    def __neg__(f): 
        return ScalarField(-f.v, f.a, f.theta_grid)

    def __pow__(f, scalar): 
        return ScalarField(f.v**scalar, f.a*scalar, f.theta_grid)
    
    def norm(f):
        return np.abs(f.v)
    
#units: 
# dist = 1 AU
# time = 1 yr
# mass = 1 M_sun
class units:
    m  = 1/149597870700 # AU per meter
    s  = 1/31556926     # yr per s
    kg = 1/(1.989 * 1e30)# msun per kg
    
class constants:
    AU = 149597870700 * units.m
    yr = 31556926*units.s
    Msun = (1.989 * 1e30)*units.kg
    G = 6.6743 * 1e-11 * units.m**3 * units.kg**-1 *  units.s**-2
    R = 1*AU
    GM = G*Msun
    c = 299_792_458 * units.m / units.s
    

class Simulation:
    def __init__(self, theta_grid, cs, Lambda_a_0, Rm_0, lambda_, rho, v, B):
        # set prescibed fields eta_O, eta_H, eta_A, c_s
        # and initial conditions rho, v, B
        """Initialize simulation.

        Parameters:
        theta_grid -- the theta grid to simulate on
        h -- disc geometrical thickness
        imag -- the imaginary part (default 0.0)
        """
        self.theta_grid = theta_grid
        z = np.arctan(self.theta_grid-np.pi/2)
        v_k = ScalarField(np.sqrt(constants.GM), -1/2, theta_grid)
        omega_k = v_k
        self.cs = ScalarField(cs, -1/2, theta_grid)
        h = ScalarField(cs[len(cs)//2], -1/2, theta_grid)
        rho_0 = rho.v[len(rho.v)//2]
        Lambda_A = ScalarField(Lambda_a_0*np.exp((z/(lambda_*h.v))**4), -1, theta_grid)
        Rm = ScalarField(Rm_0*np.exp((z/(lambda_*h.v))**4)*(rho_0/rho.v), -2, theta_grid)
        self.eta_O = omega_k*h*h/Rm
        self.eta_H = ScalarField(0, -3/2 + 2, theta_grid)
        V_A = (B*B/rho)**0.5 #find out what mu0 is in these units
        self.eta_A = V_A**2/(omega_k*Lambda_A)
    
    def J(self, B):
        return B.curl()*(constants.c/(4*np.pi))
    
    def epsilon(self, v, B):
        b_unit = B.unit()[1:-1]
        J = self.J(B)
        return -v.cross(B)[1:-1] + (
              self.eta_O[1:-1] * J \
            + self.eta_H[1:-1] * J.cross(b_unit) \
            - self.eta_A[1:-1] * J.cross(b_unit).cross(b_unit)
        )*(4*np.pi/constants.c)
    
    def get_rho_dt(self, rho, v):
        return -(rho[1:-1] * v[1:-1]).div()
    
    def get_rho_v_dt(self, rho, v, B):
        grav_force_field = VectorField(*np.reshape([constants.GM,0,0], (3,1)).repeat(len(rho.v)-4, 1), -2, thetas)   
        #print((v[2:-2]*(rho[1:-1]*v[1:-1]).div()).v_r.min())
        #print(v[1:-1].direction_deriv(rho[1:-1]*v[1:-1]).v_r.min())
        return (-(v[2:-2]*(rho[1:-1]*v[1:-1]).div() + v[1:-1].direction_deriv(rho[1:-1]*v[1:-1]))
            - (rho[1:-1]*self.cs[1:-1]**2).grad()
            #+ self.J(B[1:-1]).cross(B[2:-2])*(1/constants.c) 
            - grav_force_field * rho[2:-2]
        )
    
    def get_B_dt(self, v, B):
        return -self.epsilon(v, B).curl()

@numba.njit
def pad(values, amount, constant=True):
    res = np.empty(len(values) + 2*amount)
    res[amount:-amount] = values
    if constant:
        res[:amount] = values[0]
        res[-amount:] = values[-1]
    else:
        dy_lower = values[1] - values[0]
        dy_upper = values[-1] - values[-2]
        for i in range(amount):
            res[amount-1-i] = values[0] - dy_lower * (1+i)
            res[-(amount-i)] = values[-1] + dy_upper * (1+i)
    return res

# A divergence-free B-field uniquely sets B_r
# here, we also expand with 3 on each side, such that Br is expanded with 2 on each side
def infer_Br(B_theta, theta_grid):
    expanded_theta = pad(theta_grid, 3, False)
    expanded_B     = pad(B_theta, 3, False)
    expanded_Br = -diff(expanded_B*np.sin(expanded_theta), theta_grid[1]-theta_grid[0]) \
            /(np.sin(expanded_theta[1:-1]) * (-5/4 + 2))
    return expanded_Br
    
def un_flatten(x, theta_grid, cs_centre):
    expanded_theta = pad(theta_grid, 2, False) #expand with 2 on each side
    rho_data, vr_data, vtheta_data, vphi_data, Btheta_data, Bphi_data = np.split(x, 6)
    rho = ScalarField(pad(rho_data**2/np.sum(rho_data**2), 2), -3/2, expanded_theta)
    v = VectorField(*[pad(v, 2) for v in [vr_data, vtheta_data, vphi_data]],
              -1/2, expanded_theta)
    beta = 1e9
    B_target = np.sqrt(8*np.pi*(rho_data[len(rho_data)//2]**2)*cs_centre**2/beta)#1 should be cs
    B_multiplier = B_target/Btheta_data[len(rho_data)//2]
    #print(v*rho)
    #print(v*v)
    B = VectorField(infer_Br(Btheta_data, theta_grid), *[pad(B, 2, False) for B in [Btheta_data, Bphi_data]], -5/4, expanded_theta)*B_multiplier
    return rho, v, B
    
def get_stationary_residual(x, theta_grid):
    cs, Lambda_a_0, Rm_0, lambda_ = np.ones(len(theta_grid)+4)*0.21, 1, np.inf, 3 #HOW TO??
    rho, v, B = un_flatten(x,theta_grid, cs[len(cs)//2])
    sim = Simulation(rho.theta_grid, cs, Lambda_a_0, Rm_0, lambda_, rho, v, B)
    return sim.get_rho_dt(rho, v), sim.get_rho_v_dt(rho, v, B),sim.get_B_dt(v, B)
    
def get_resudual_error(residual):
    return np.sum([v.norm() for v in residual])

def residual_to_neg_gradient(rho_res, rho_v_res, B_res, rho, v):
    vres = (rho_v_res-v[2:-2]*rho_res)/rho[2:-2]
    return np.concatenate([rho_res.v, vres.v_rho, vres.v_theta, vres.v_phi, B_res.v_theta, B_res.v_phi])
```

```python
N=1000
cs = 1e3 * units.m/units.s
rho_0 = 1e-6 * units.kg/units.m**3
h = 0.5
dtheta = np.pi/(N+6)
thetas = np.linspace(0.15, np.pi-0.15, N, endpoint=True)
z = np.arctan(thetas - np.pi/2)
rho_init = rho_0*np.exp(-z**2/(2*h**2))*np.sin(thetas)**(3/2)
v_init = np.concatenate([np.zeros(N), np.zeros(N), np.zeros(N) + 2*np.pi*constants.AU/constants.yr])
B_init = np.concatenate([-np.sin(thetas), np.zeros(N)]) # only specify theta and phi
x_init = np.concatenate([rho_init, v_init, B_init])#+ np.random.rand(7*N)*1e-5
```

```python
import scipy.optimize
np.seterr("warn", under="ignore")
def get_error(x): 
    err = get_resudual_error(get_stationary_residual(x, thetas))
    #print(count, np.sum(err[0]**2))
    return err

count=0
def print_iter(x):
    global count
    count += 1
    err = get_resudual_error(get_stationary_residual(x, thetas))
    print(count, err)
    
%load_ext snakeviz
#%snakeviz get_error(x_init)
res = scipy.optimize.minimize(get_error, x_init, options={"disp":True, "maxiter":50}, callback=print_iter)
#res
```

```python
def plot_system(rho, v, B, axes):
    rho_ax, v_ax, B_ax = axes
    rho_ax.plot(rho.theta_grid, rho.v, color="k")
    rho_ax.set_title("rho")
    v_ax.plot(v.theta_grid, v.v_r, color="blue")
    v_ax.plot(v.theta_grid, v.v_theta, color="red")
    v_ax.plot(v.theta_grid, v.v_phi, color="green")
    v_ax.set_title("v")
    v_ax.legend(["r", "theta", "phi"])
    B_ax.plot(B.theta_grid, B.v_r, color="blue")
    B_ax.plot(B.theta_grid, B.v_theta, color="red")
    B_ax.plot(B.theta_grid, B.v_phi, color="green")
    B_ax.set_title("B")
    B_ax.legend(["r", "theta", "phi"])
    
fig, axes = plt.subplots(1, 3, figsize=(20,6))
system = x_init
for it in range(20):
    state = un_flatten(system, thetas, 0.21)
    residual = get_stationary_residual(system, thetas)
    gradient = residual_to_neg_gradient(*residual, *state[:2])
    system = system + 0.001*gradient

plot_system(*state, axes)
```
