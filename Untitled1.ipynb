```python
import pandas as pd
import re
import os
from spherical_fields import *
from mhd_system import *
from settings import Settings
import matplotlib.pyplot as plt

def read_lesur(filename):
    global data
    global theta
    data = pd.read_csv(filename, skiprows=6, delimiter="\t+", engine='python')
    data.columns = [x.split(" ")[0] for x in data.columns]
    theta = data.theta.to_numpy()
    global B
    global rho
    global v
    rho = ScalarField(data.rho.to_numpy(), -3/2, theta)
    v = VectorField(data.vr.to_numpy(), data.vtheta.to_numpy(), data.vphi.to_numpy(), -1/2, theta)
    B = VectorField(data.Br.to_numpy(), data.Btheta.to_numpy(), data.Bphi.to_numpy(), -5/4, theta)
    cs = ScalarField(np.sqrt((data.P/data.rho).to_numpy()), 0, theta, prescription="cylindrical")
    beta, Rm_0, Lambda_a_0 = [float(x) for x in re.findall("=(inf|[^-]+)", filename)]
    lambda_ = 3
    settings = Settings(cs, Lambda_a_0, Rm_0, lambda_, beta = 0, Sigma = 0)
    return mhd_system(settings)


for file in os.listdir("../PPDwind"):
    if ".dat" in file:
        sim = read_lesur("../PPDwind/" + file)
        #plt.plot(theta,cs2.v)
        plt.semilogy(theta,rho.v)
        #plt.xlim(np.pi/2-0.01, np.pi/2+0.01)
        break
```

```python
plt.plot(sim.get_B_dt(rho, v, B).v_r, label="r")
plt.plot(sim.get_B_dt(rho, v, B).v_theta, label="theta")
plt.plot(sim.get_B_dt(rho, v, B).v_phi, label="phi")
plt.legend()
plt.yscale("symlog")
```
